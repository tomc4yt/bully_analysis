```python
import nltk
import pandas as pd
import numpy as np
import os as os
```

```python
bully_df = pd.read_csv('data/QualStudy1.csv', encoding='cp1252')
```

```python
bully_df.head()
```

```python
# Remove Stop Words
stop_words = nltk.corpus.stopwords.words("english")

bully_paragraphs = bully_df.Story.tolist()
tokenized_text = {}

# Tokenize Text
for i in range(0,len(bully_paragraphs)):
    tokenized_text[i] = {}
    tokenized_sentences = nltk.sent_tokenize(bully_paragraphs[i])
    for j in range(0, len(tokenized_sentences)):
        
        tokenized_text[i][j] = nltk.word_tokenize(tokenized_sentences[j])
        
        # Clean data
        for word in tokenized_text[i][j]:
            
            # Remove stop words
            if word in stop_words:
                tokenized_text[i][j].remove(word)

```
